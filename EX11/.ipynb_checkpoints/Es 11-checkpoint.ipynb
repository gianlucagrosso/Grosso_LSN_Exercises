## Machine Learning(ML): regressione con reti neurali (NN)

Il **Machine Learning** può essere introdotto come un sottocampo dell'intelligenza artificiale (**AI**) con l'obiettivo di sviluppare algoritmi in grado di imparare dai dati in modo **automatico**. 

La principale premessa è che il ML costruisce algoritmi che possano ricevere dati in input per **predire**, usando analisi statistiche, outputs che siano costantemente aggiornati mano a mano che nuovi dati di input diventano disponibili.

IL ML si divide in tre ampie **categorie**. In questa esercitazione l'attenzione sarà rivolta al **Supervised Learning (SL)**: esso impara da dati **etichettati**, con l'obiettivo di approssimare così bene la funzione che mappa i dati di input(X) in quelli di output(Y), disponibili in ingresso, in modo da saper predire nuove variabili di output assegnati nuovi dati di input, diversi da quelli sul quale si era precedentemente *allenato*. 

I task di apprendimento più comuni nel contesto del SL sono:

1. **Classificazione**: dove la variabile di output rappresenta una categoria.

2. **Regressione**: dove la variabile di output è un numero reale, per esempio, un valore di una funzione fittata.

In questa esercitazione viene affrontato il secondo tipo di apprendimento, nel tentativo di imparare a fittare funzioni sfruttando la tecnica più potente  e ampiamente usata per l'apprendimento supervisionato, quella delle **reti neurali(NN)**, imitando il funzionamento del neurone biologico.

## Esercizio 11.1

Nel primo esercizio viene richiesto di migliorare i risultati ottenuti (sul notebook contenente le consegne) per un fit di un modello lineare $f(x)=2x+1$ modificando alcuni parametri.

Un problema nell'ambito del SL deve partire da un **DataSet ($X,Y$)**, dove X sono un insieme di variabili indipendenti e Y un insieme di variabili dipendenti. Quest'ultimo verrà diviso in due gruppi mutuamente esclusivi: un dataset di **train** (allenamento) e uno di **test** (o validazione). La maggior parte dei dati viene inserità nel gruppo di train.  

Un altro ingrediente indispensabile è il **modello**: una funzione $f(X,p)$ dei parametri p che mappa le X nelle Y e che viene impiegata per effettuare delle predizioni sugli output.

Nell'esercizio si sfrutta una NN, che dipenderà dai suoi parametri, **pesi** e **biases**, per fittare il modello. Di conseguenza la NN può modellizzare una numerosa classe di funzioni.

 Per apprendere i *parametri ottimali* della rete, quest'ultima viene allenata sul dataset di training e ne viene testata la sua efficacia e la sua performance sul dataset(diverso) di validazione perchè l'obiettivo finale consiste nel trovare l'archittetura migliore della rete che sia in grado di rappresentare il modello e di elaborare predizioni accurate a partire da nuovi dati di input mai visti.

L'ultimo elemento fondamentale è la funzione costo $C(Y;f(X,p))$ che permette di giuidicare quanto bene funzioni il modello, ossia quanto le sue predizioni siano compatibili con le variabili dipendenti di output disponibili.

Usando una rete neurale, ciò significa trovare una funzione costo che misuri le performances predittive dell'algoritmo confrontando le predizioni ${y_i}^{pred}$ con i valori veri ${y_i}^{true}$ disponibili nel dataset di input.

Nello specifico, l'obiettivo è quello di trovare i pesi e i parametri ottimali della rete che minimizzano la funzione costo scelta.

Definite quindi una funzione costo e un'archittetura di una rete neurale, si procederà allenando il modello e ottimizando la funzione costo con degli algoritmi di ottimizzazione gradient-based tra i quali quello più frequentemente utilizzato è lo Stochastic Gradient Descent(SGD)


Tornando all'esercizio l'obiettivo è quello di allenare una NN con un singolo layer e un singolo neurone (sufficiente per modellizzare un fit lineare) disponendo di un dataset di input ($N_{train} = 500$, $N_{valid} = 50$) che viene leggermente deviato dal comportamento ideale della funzione $f(x)$ con un **rumore gaussiano** ($y_i= f(x_i) + \eta_i $) per trovare i parametri migliori della rete che ottimizzino la funzione costo (inizialmente la **mean square-error(MSE)**),usando come ottimizzatore lo SGD, per effettuare predizioni quanto più accurate possibili a partire da nuovi dati di input. Il tentativo di fittare il modello usando il gruppo di dati di train viene ripetuto per un certo numero di epoche ($N_{epoch} = 30$), dove ogni epoca si conclude quando l'intero gruppo del training dataset, suddiviso in *minibatches* ($N_{batch} = 32$), è stato esplorato per allenare la rete.

Si osserva, come mostrato di seguito, che si ottiene un fit migliore dei punti distribuiti lungo la retta $f(x)=2x+1$:

1) innanzitutto aumentando il numero di punti utilizzati per l'apprendimento della rete neurale di un fattore 10 ($N_{train} = 5000$, $N_{valid} = 500$) mantenendo inalterato il numero di epoche ($N_{epoch} = 30$) settando un rumore gaussiano $\sigma = 0.3$

2) oppure mantenendo invariata la dimensione del dataset di input ma aumentando in modo considerevole il tempo di apprendimento ($N_{epoch} = 200$). In questo caso bisogna stare attenti di non cadere nel pericolo dell'**overfitting** dei dati di input, infatti, se l'allenamento è portato avanti per un numero di epoche troppo elevato il rischio è di inziare a fittare anche il rumore con il quale è stata "sporcata" la funzione lineare. Questo controllo si può effettuare con la cross-validation verifcando l'andamento (in funzione delle epoche), rispettivamente, dell'$E_{out}$ (errore di generalizzazione), ossia il valore minimo della funzione costo per il miglior modello fittato sui dati di validazione, e dell'$E_{in}$, il minimo sui dati di training. Quando l'errore del campione di validazione inizia a crescere, invertendo il suo andamento descrescente, significa che la NN sta overfittando. Tale criterio è detto **Early-Stopping method**. 

3) Infine si osserva che aumentando il valore del disturbo gaussiano $\sigma$ per ottenere dei risultati soddisfacenti bisogna aumentare il numero di dati di apprendimento, altrimenti il rischio è di confondere l'andamento lineare originario con il rumore crescente.


### 1) Aumenta la dimensione del dataset di input; invariato il numero di epoche; rumore gaussiano $\sigma = 0.3$

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

sigma = 0.3 # gaussian noise

# basic parameters
N_train=5000
N_valid=500
N_predict=500
N_epoch=30
```

```python
import numpy as np
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, N_train)
x_valid = np.random.uniform(-1, 1, N_valid)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target points')

plt.scatter(x_valid, y_valid, color='r', label='validation points')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over N_epoch epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=N_epoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
plt.figure()
x_predicted = np.random.uniform(-1, 1, N_predict)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r', label='predicted points')
plt.plot(x_valid, y_target, label='target fit')
plt.grid(True)
plt.legend()
plt.show()
```

## Esercizio 11.2

Nel secondo esercizio è richiesto di estendere la *potenzialità* dello strumento delle reti neurali per ottenere un fit *ragionevole* del seguente **polinomio** di terzo odine:
$$
f(x)=4-3x-2x^2+3x^3
$$
per $x \in [-1,1]$.

Per trovare il fit migliore risulterà necessario fare esperienza variando i seguenti parametri e controllando i risultati ottenuti:

- Il numero di **layer** che compongono la NN (la **profondità** della rete): ogni rete neurale è infatti composta da un certo numero di layer di tre possibili tipi ( input,nascosti e output) che nelle NN feedforward sono connessi solo con il successivo.


- Il numero di **neuroni** in ciascun layer(l'**ampiezza** della rete): insieme al numero di layer costituiscono l'**architettura** della rete. La configurazione dei layer di input e di output dovrà adattarsi al dataset (X,Y).


- la **funzione di attivazione**: una specifica funzione, di solito non lineare, della variabile $ z = \vec{w} \cdot \vec{x} + w_0 $ (dove $\vec{w}$ sono i pesi, $\vec{x}$ gli input e $w_0$ il bias ), che è responsabile della risposta in output di ogni nodo di una rete e assegnata ad ogni layer. 


- L'**ottimizzatore**: l'algoritmo che ottimizza i pesi e i bias delle NN durante la procedura di allenamento con lo scopo di minimizzare la funzione costo.


- La **funzione costo**: la funzione da minimizzare durante l'ottimizzazione che fornisce un'indicazione di quanto bene stia funzionando il modello della rete con i parametri attuali.


Nei tentativi, che sono riportati e commentati di seguito, i layer di input e di output sono sempre costituiti da un solo neurone poichè è presente una sola variabile sia in entrata che in uscita. 

### Tentativo 1: Architettura

Il primo esperimento consiste nell'allenare una rete neurale progettata con un solo layer **denso nascosto** di ampiezza $N_{neu} = 20 $ unità con funzione di attivazione RELU (la più comune) per $ N_{epoch} = 50 $ usando l'SGD come ottimizzatore e la MSE come funzione costo e come misura della precisone (la cosiddetta metrica). I dati su cui si allena la rete e su cui viene validata sono $N_{train} = 10^4$, $N_{valid} = 10^3$. 

Tale rete, rispetto all'esercizio precedente, deve essere caratterizzata da una complessità maggiore per fittare al meglio le oscillazioni che interessano una polinomio del terzo ordine. Con un numero di parametri superiore, la rete realizza numerose funzioni caratteristiche da sommare.

**N.B.**: il layer di input ed output, per ora, non possiedono alcuna funzione di attivazione per avere i dati in entrata ed in uscita "puliti".

```python
def my_f(x):
  # target parameters of f(x) =  d + c*x + b*x*x +a*x*x*x
  a = 3 # x^3 coefficient
  b = -2 # x^2 coefficient
  c = -3 # x coefficient
  d = 4 # pure coefficien 
  return d+c*x+b*x*x+a*x*x*x

sigma = 0.2 # gaussian noise
#basic parameter
N_train=10000
N_valid=1000
N_predict=1000
N_epoch=50

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, N_train)
x_valid = np.random.uniform(-1, 1, N_valid)
x_valid.sort()
y_target = my_f(x_valid) # ideal (target) parabolic function

y_train = np.random.normal(my_f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(my_f(x_valid), sigma)
```

```python
# plot training and target dataset
plt.plot(x_valid, y_target, label='target points')
plt.scatter(x_valid, y_valid, color='r', label='validation points')
plt.grid(True)
plt.legend()
plt.show()
```

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))

from keras import optimizers, losses, metrics

sgd = tf.optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.4, nesterov=True)

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 50 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=N_epoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, N_predict)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r', label='predicted points')
plt.plot(x_valid, y_target, label='target fit')
plt.grid(True)
plt.legend()
plt.show()
```

- Il numero dei parametri della NN sono 63, una rete ancora non troppo complessa. I risultati infatti non si sono rivelati soddisfacenti, con andamenti troppo "spezzati" e con scarso potere predittivo per $x \in [0.25, 1]$.

### Tentativo 4: Architettura
Nel quarto tentativo, date le precedenti considerazioni, si prova ad archittetare una rete con più strati nascosti, cercando di mantenere lo stesso numero di parametri (circa 363). La struttura scelte prevede 10 neuroni per il primo strato nascosto, 15 per il secondo e 10 per il terzo. Il resto dei parametri e delle scelte rimangono invariati.

### Esercizio 11.3

L'ultimo esercizio richeide di estendere ulteriormente il modello della NN per fittare in modo ragionevole una semplice funzione **trigonometrica bidimensionale**
$$
f(x,y) = \sin(x^2+y^2)
$$ 
nel dominio $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

In base alle osservazioni e all'esperienza acquisita con lo scorso esercizio sono state adottate le seguenti scelte:

- $N_{train} = 10^4$ e $N_{valid} = 10^3$.

- $N_{epoch} = 200$ e rumore gaussiano $\sigma = 0.2$

- Una rete neurale con la struttura $NN(2,70,45,30,1)$ con profondità pari a 3 e numero di neuroni per layer nascosto che decrescono mano a a mano che le connessioni procedono. Il layer di input ha due neuroni di input perchè due sono le variabili in ingresso (funzione 2D).

- Funzione di attivazione RELU per gli strati nascosti, assente per il layer di input e tanh per il layer di output per avere dati in uscita normalizzati nell'intervallo [-1,1] corrispondente al dominio del modello da fittare $f(x,y)$.

- Ottimizzatore l'Adadelta

- La funzione costo (e la metrica) la MSE.




```python
# definition f(x) = sin(x*x + y*y)
def sin2d(x): # "2D function"
    return np.sin(x[:,0]**2+x[:,1]**2)

sigma = 0.2 # gaussian noise
#basic parameter
N_train=10000
N_valid=1000
N_predict=1000
N_epoch=200

# generate training inputs
np.random.seed(0)
x_train = np.zeros((N_train, 2))  # input matrix (x,y)
z_train = np.zeros((N_train, 1))  # z = f(x,y)

x_valid = np.zeros((N_valid, 2))
z_valid = np.zeros((N_valid, 1))

x_train[:,0] = np.random.uniform(-3./2., 3./2., N_train)
x_train[:,1] = np.random.uniform(-3./2., 3./2., N_train)

x_valid[:,0] = np.random.uniform(-3./2., 3./2., N_valid)
x_valid[:,1] = np.random.uniform(-3./2., 3./2., N_valid)

z_target= sin2d(x_valid) # ideal (target) trigonometric function

z_train = np.random.normal(sin2d(x_train), sigma)
z_valid = np.random.normal(sin2d(x_valid), sigma) # actual measures from which we want to guess regression parameters
```

```python
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# plot training and target dataset


fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_valid[:,0], x_valid[:,1],z_valid, label='target points')
ax.scatter(x_valid[:,0], x_valid[:,1],z_target, c='red', label='validation points')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
#ax.view_init(10, 30)
plt.show()
```

```python
model = tf.keras.Sequential()

model.add(Dense(2, input_shape=(2,)))
model.add(Dense(70, activation='relu'))
model.add(Dense(45, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='tanh'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer=tf.keras.optimizers.Adadelta(lr=1.0, rho=0.95, epsilon=None, decay=0.0), loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 200 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=z_train, 
          batch_size=32, epochs=N_epoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, z_valid))
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, z_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
